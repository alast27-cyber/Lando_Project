The `pipeline` function from the `transformers` library makes it very easy to use pre-trained models for various tasks without needing to write much code. Here's an example for text generation:

```python
from transformers import pipeline

# Initialize a text generation pipeline
generator = pipeline('text-generation', model='distilgpt2')

# Generate text with a new prompt
result = generator("In a land far, far away,", max_length=60, num_return_sequences=1)

# Print the generated text
print(result[0]['generated_text'])
```

```python
from transformers import pipeline

# Initialize a question-answering pipeline
qa_pipeline = pipeline("question-answering")

# Provide a question and a context
question = "What is the capital of France?"
context = "Paris is the capital and most populous city of France. An important centre of finance, commerce, fashion, science, and the arts, Paris has a major influence on politics, education, entertainment, media, science, and the arts, and is one of the world's major global cities."

# Get the answer
answer = qa_pipeline(question=question, context=context)

print(f"Question: {question}")
print(f"Context: {context}")
print(f"Answer: {answer['answer']} (Score: {answer['score']:.2f})")
```

This code uses a default question-answering model to find the answer to a given question within a provided context. The output will include the extracted answer, its confidence score, and the start and end positions within the context.

```python
from transformers import pipeline

# Initialize a fill-mask pipeline
fill_masker = pipeline('fill-mask', model='distilroberta-base')

# Use a masked sentence
result_mask = fill_masker("The capital of France is [MASK].")
print("Fill-mask analysis for 'The capital of France is [MASK].':\n", result_mask)

result_mask_2 = fill_masker("Paris is the [MASK] of France.")
print("\nFill-mask analysis for 'Paris is the [MASK] of France.':\n", result_mask_2)
```

This code uses a pre-trained RoBERTa model (`distilroberta-base`) to predict the most likely words to fill the `[MASK]` token in the given sentences. It will output a list of possible words along with their confidence scores.

```python
from transformers import pipeline

# Initialize a sentiment analysis pipeline
sentiment_analyzer = pipeline('sentiment-analysis')

# Analyze a positive sentence
result_positive = sentiment_analyzer("I love using Hugging Face Transformers!")
print("Positive sentence analysis:", result_positive)

# Analyze a negative sentence
result_negative = sentiment_analyzer("I hate this cold weather.")
print("Negative sentence analysis:", result_negative)
```

This code will download a default sentiment analysis model and then use it to classify the sentiment of the provided sentences as either 'POSITIVE' or 'NEGATIVE', along with a confidence score.
