The `pipeline` function from the `transformers` library makes it very easy to use pre-trained models for various tasks without needing to write much code. Here's an example for text generation:

```python
from transformers import pipeline

# Initialize a text generation pipeline
generator = pipeline('text-generation', model='distilgpt2')

# Generate text with a new prompt
result = generator("The quick brown fox jumps over the lazy dog,", max_length=50, num_return_sequences=1)

# Print the generated text
print(result[0]['generated_text'])
```

This code downloads a small, pre-trained GPT-2 model (`distilgpt2`) and uses it to complete the given prompt. You can replace `'text-generation'` with other tasks like `'sentiment-analysis'`, `'fill-mask'`, `'question-answering'`, etc., and choose different models accordingly.
